Q1 asks to implemet a function that takes two inputted points and finds the Euclidean distance between them. The following code asks the user for those inputs breaking it down into x and y values for each point. The x and y values are computed into the Euclidean distance using the math module and the result is printed.

```python
valuex1=int(input("What is your first x value?:"))
valuex2=int(input("What is your second x value?:"))
valuey1=int(input("What is your first y value?:"))
valuey2=int(input("What is your second y value?:"))

def EucDistance(x,y,x1,y1):
    xsquare=(x-x1)**2
    ysquare=(y-y1)**2
    distance= round((math.sqrt(xsquare + ysquare)),3)
    print("the euclidean distance for (", x,",", y,")"," and ","(", x1,",", y1,")" ," is:", distance)

EucDistance(valuex1,valuey1,valuex2,valuey2)
```

Q3 asks to create a function that returns the spherical distance beween two points. The following code asks the user for their input on two locations broken up into longitudes and latitudes. In a function defined as "SphDist" takes those inputs and plugs them into the haversine equation, after converting the values to radians, and uses the math module. Upon request, the points for (-33, 151) and (40,-88) are inserted into the function, finding their spherical distance of 14780.043 kms.

```python
lon=float(input("What is your first longitude value?:"))
lat=float(input("What is your first latitude value?:"))
lon2=float(input("What is your second longitude value?:"))
lat2=float(input("What is your second latitude value?:"))

def SphDist(long1,lati1,long2,lati2):
    radlong1=math.radians(long1)
    radlati1=math.radians(lati1)
    radlong2=math.radians(long2)
    radlati2=math.radians(lati2)
    sublong=(radlong1-radlong2)/2
    sublat=(radlati1-radlati2)/2
    a=math.sin(sublat)**2 + math.cos(radlati1) *math.cos(radlati2) *math.sin(sublong)**2
    c=2*(math.asin(math.sqrt(a)))
    SphD=round((6371.0*c),3)
    print("the spherical distance for (", lati1,",", long1,")"," and ","(", lati2,",", long1,")" ," is:", SphD,"kms")

SphDist(lon,lat,lon2,lat2)
SphDist(151,-33,-88,40)
```

Q4 asks to create two line segments and determine if they intersect using shapely. To come up with a set of lines that do intercept and a set of lines that do not, there are 3 different line segment variables. Using ".intersects" from shapely, the written code returns a boolean where "true" indicates the lines intersect. The function is used twice with the different lineStrings to have one outcome intersect and one not.

```python
line1=LineString([(1,0), (4,6)])
line2=LineString([(2,0), (3,8)])
line3=LineString([(2,0), (4,0)])
def lineInt(line, lineComp):
  if line.intersects(lineComp):
    print("These lines do intercept")
  else:
    print("These lines do not intercept")
print("Line Interception (Q4)")
lineInt(line1, line2)
lineInt(line1, line3)
```

Q5 asks to create a rectangle as a polygon with shapely and create a function to determine if a point is inside or outside of it. To set up the function, a list with vertices for the polygon were created along with two points (one that woud eventually fall in the shape and one that would not). Using ".contains" with shapely the program tests if the point is within the bounds, working similar to how the line interception did in Q4. If the ".contains" function is true then the return will indicate the point is in the polygon and the opposite if not.

```python
coords=[(0,0), (5,0), (5,10), (0,10)]
point1= Point(1,1)
point2=Point(13,5)
rectangle= Polygon(coords)

def pointContains(shape, point, coordinates):
  if shape.contains(point):
    print( point," is inside of shape with vertices at", coordinates)
  else:
    print( point," is outside of shape with vertices at", coordinates)

print("Point in a polygon (Q5)")
pointContains(rectangle, point1, coords)
pointContains(rectangle, point2, coords)
```
