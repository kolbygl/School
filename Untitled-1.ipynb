Q1 asks to implemet a function that takes two inputted points and finds the Euclidean distance between them. The following code asks the user for those inputs breaking it down into x and y values for each point. The x and y values are computed into the Euclidean distance using the math module and the result is printed.

```python
valuex1=int(input("What is your first x value?:"))
valuex2=int(input("What is your second x value?:"))
valuey1=int(input("What is your first y value?:"))
valuey2=int(input("What is your second y value?:"))

def EucDistance(x,y,x1,y1):
    xsquare=(x-x1)**2
    ysquare=(y-y1)**2
    distance= round((math.sqrt(xsquare + ysquare)),3)
    print("the euclidean distance for (", x,",", y,")"," and ","(", x1,",", y1,")" ," is:", distance)

EucDistance(valuex1,valuey1,valuex2,valuey2)
```

Q3 asks to create a function that returns the spherical distance beween two points. The following code asks the user for their input on two locations broken up into longitudes and latitudes. In a function defined as "SphDist" takes those inputs and plugs them into the haversine equation, after converting the values to radians, and uses the math module. Upon request, the points for (-33, 151) and (40,-88) are inserted into the function, finding their spherical distance of "".

```python
lon=float(input("What is your first longitude value?:"))
lat=float(input("What is your first latitude value?:"))
lon2=float(input("What is your second longitude value?:"))
lat2=float(input("What is your second latitude value?:"))

def SphDist(long1,lati1,long2,lati2):
    radlong1=math.radians(long1)
    radlati1=math.radians(lati1)
    radlong2=math.radians(long2)
    radlati2=math.radians(lati2)
    sublong=(radlong1-radlong2)/2
    sublat=(radlati1-radlati2)/2
    a=math.sin(sublat)**2 + math.cos(radlati1) *math.cos(radlati2) *math.sin(sublong)**2
    c=2*(math.asin(math.sqrt(a)))
    SphD=round((6371.0*c),3)
    print("the spherical distance for (", lati1,",", long1,")"," and ","(", lati2,",", long1,")" ," is:", SphD,"kms")

SphDist(lon,lat,lon2,lat2)
SphDist(151,-33,-88,40)
```
